## RANDOM FOREST

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
selected_columns = ['Age', 'Gender', 'Years at Company', 'Job Role', 
                    'Number of Promotions', 'Education Level', 
                    'Number of Dependents', 'Job Level', 
                    'Company Tenure', 'Remote Work', 'Attrition','Marital Status','Distance from Home']

# Crear un nou dataframe amb les columnes seleccionades
df_selected = df_clean[selected_columns]
```

```python
# Paso 2: Submuestreo aleatorio a 15,000 muestras
df_sampled = df_selected.sample(n=2000, random_state=42)
```

```python
x = df_sampled.drop(columns=['Attrition'])
y = df_sampled['Attrition']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.28, random_state = 42)
```

```python
X_train.shape, X_test.shape
```

```python
# Escalar los datos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# 4. Entrenamiento del modelo Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train_scaled, y_train)
```

```python
# 5. Evaluación del modelo
y_pred = rf_model.predict(X_test_scaled)
```

```python
print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred))
print("Matriz de Confusión:\n", confusion_matrix(y_test, y_pred))
```

El model Random Forest té una precisió global del 66%, amb un millor rendiment per identificar els empleats que marxen (classe 1) amb una precisió del 72% i un recall del 70%. Per a la classe 0 (Podem observar que el model és més efectiu a l'hora de predir la classe 1 (empleats que marxen), mentre que és menys precís per a la classe 0 (empleats que es queden). Això es deu al desbalanceig de la variable objectiu, on la classe 1 és més representada que la classe 0. Per millorar el rendiment i ajustar el model, podríem utilitzar el paràmetre class_weight='balanced' per equilibrar la importància de les dues classes, de manera que el model presti més atenció als casos menys representats.empleats que es queden), la precisió és més baixa (59%), i el recall és de 62%.

Podem observar que el model en aquest cas es mes efectiu al fer prediccions quan es 1, en canvi quan es 0 el model es menys efectiu, ja que com hem observat 
la base de dades en la variuable objectiu els seus dos camps unics entre ells estan una mica desbalancejats, per millorar y ajustar el problema podriem 
utilitzar el parametre weight en balanced.


```python
importances = rf_model.feature_importances_
feature_importance_df = pd.DataFrame({'Feature': X_train.columns, 'Importance': importances})
print("\nFeature Importances:\n", feature_importance_df.sort_values(by='Importance', ascending=False))
```

## Random amb ajustament parametres

```python
# 4. Entrenamiento del modelo Random Forest
rf_model = RandomForestClassifier(max_depth=3,class_weight='balanced',random_state=42)
rf_model.fit(X_train_scaled, y_train)
y_pred = rf_model.predict(X_test_scaled)
print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred))
print("Matriz de Confusión:\n", confusion_matrix(y_test, y_pred))
```

Després d'ajustar els paràmetres, el model ha mostrat millores en la precisió, recall i F1-score, especialment per a la classe 1 (empleats que marxen). L'accuracy global ha augmentat del 66% al 68%. La matriu de confusió també mostra menys errors en la classificació de les dues classes, indicant una millor equilibració en les prediccions.

## RANDOM 2

```python
rf_model2 = RandomForestClassifier(random_state=42)

param_grid = {
    'n_estimators': [50, 100, 200],          # Número de árboles
    'max_depth': [None, 10, 20, 30],        # Profundidad máxima del árbol
    'min_samples_split': [2, 5, 10],        # Número mínimo de muestras para dividir un nodo
    'min_samples_leaf': [1, 2, 4],          # Número mínimo de muestras en una hoja
    'bootstrap': [True, False],
    'class_weight': ['balanced']
}

# 4. Configurar el GridSearchCV
grid_search = GridSearchCV(estimator=rf_model2, 
                           param_grid=param_grid, 
                           cv=5,                  # Validación cruzada con 5 particiones
                           scoring='accuracy',    # Métrica a optimizar
                           verbose=2,             # Mostrar progreso
                           n_jobs=-1)             # Usar todos los núcleos disponibles

# 5. Ejecutar la búsqueda de hiperparámetros
grid_search.fit(X_train, y_train)

# 6. Mostrar los mejores parámetros encontrados
print("Mejores hiperparámetros:", grid_search.best_params_)

# 7. Evaluar el modelo con los mejores parámetros en los datos de prueba
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
```

```python
print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred))
print("Matriz de Confusión:\n", confusion_matrix(y_test, y_pred))
```
